```python
def dfs_recur(G, v, visitado=None, contador=0):
    """
    Realiza un recorrido DFS sobre un grafo G

    Parámetros:
        G (networkx.Graph o DiGraph): Grafo de networkx.
        v: Nodo desde el cual comienza el recorrido DFS.
        visitado (dict, opcional): Diccionario para marcar nodos visitados.
        contador (int, opcional): Contador de nodos visitados.

    Devuelve:
        int: Cantidad de nodos visitados en el recorrido DFS.
    """
    if visitado == None: visitado = {}
    visitado[v] = True
    contador += 1
    lista_visitados = [v]

    for vecino in G.neighbors(v):
        if not visitado.get(vecino,False):
            contador,lista = dfs_recur(G, vecino, visitado, contador)
            lista_visitados += lista
    
    return contador, lista_visitados

def dfs(G, v, visitado=None, contador=0):
    """
    Realiza un recorrido DFS sobre un grafo G

    Parámetros:
        G (networkx.Graph o DiGraph): Grafo de networkx.
        v: Nodo desde el cual comienza el recorrido DFS.
        visitado (dict, opcional): Diccionario para marcar nodos visitados.
        contador (int, opcional): Contador de nodos visitados.

    Devuelve:
        int: Cantidad de nodos visitados en el recorrido DFS.
    """
    visitados = {}
    pila = [v]
    contador = 0
    lista_visitados = []


    while pila:
        v = pila.pop()

        if not visitados.get(v,False):
            visitados[v] = True

            contador += 1
            lista_visitados.append(v)

            for vecino in G.neighbors(v):
                pila.append(vecino)
    
    return contador, lista_visitados

```

### Ejercicio 6

Escribir una función que reciba un grafo no dirigido y retorne una lista con todas los componentes conexas del grafo. Debe usar DFS.

```python
def componentes_conexas(G):
    """
    Retorna una lista de componentes conexas de un grafo no dirigido usando DFS.

    Parámetros:
        G (networkx.Graph): Grafo no dirigido.

    Devuelve:
        list: Lista de listas, cada una contiene los nodos de una componente conexa.
    """
    visitado = {nodo: False for nodo in G.nodes}
    componentes = []

    def dfs_componente(v, actual):
        visitado[v] = True
        actual.append(v)
        
        for vecino in G.neighbors(v):
            if not visitado[vecino]:
                dfs_componente(vecino, actual)


    for nodo in G.nodes:
        if not visitado[nodo]:
            comp = []
            dfs_componente(nodo, comp)
            componentes.append(comp)
    return componentes
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G_ejemplo = nx.Graph()
G_ejemplo.add_edges_from([
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'C'),
    ('D', 'E'),
    ('F', 'G'),
    ('G', 'H'),
    ('H', 'F')
])

componentes = componentes_conexas(G_ejemplo)
print("Componentes conexas:", componentes)

# Graficar el grafo y resaltar cada componente con un color diferente
colors = ['lightblue', 'lightgreen', 'salmon', 'orange', 'violet', 'yellow']
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G_ejemplo)  # Generar posiciones para todos los nodos
for i, comp in enumerate(componentes):
    nx.draw_networkx_nodes(G_ejemplo, pos, nodelist=comp, node_color=colors[i % len(colors)], node_size=400)
nx.draw_networkx_edges(G_ejemplo, pos, edge_color='gray')
nx.draw_networkx_labels(G_ejemplo, pos)
plt.title("Componentes conexas en el grafo")
plt.axis('off')
plt.show()
```

### Ejercicio 7

Proponer un algoritmo que, dado un grafo dirigido, detecte si existe algún ciclo utilizando DFS.  
> Pista: Durante el recorrido DFS, si se encuentra una arista hacia un nodo que está actualmente en la pila de recursión (visitado pero no finalizado), se ha detectado un ciclo.  
Explicar cómo se identifican y procesan las aristas de retorno (back edges) en el recorrido.

```python
def tiene_ciclo_dfs(G):
    """
    Detecta si un grafo dirigido G tiene ciclos usando DFS.

    Parámetros:
        G (networkx.DiGraph): Grafo dirigido.

    Devuelve:
        bool: True si existe al menos un ciclo, False si es acíclico.
    """
    visitado = {nodo: False for nodo in G.nodes}
    en_recursion = {nodo: False for nodo in G.nodes}
    # Explicación:
    # En el recorrido DFS, se marca cada nodo como "en_recursion" al entrar y se desmarca al salir.
    # Si durante la exploración de vecinos se encuentra un nodo que está "en_recursion",
    # significa que existe una arista de retorno (back edge) y por lo tanto un ciclo.
    
    def dfs(v):
        visitado[v] = True
        en_recursion[v] = True

        for vecino in G.neighbors(v):
            if not visitado[vecino]:
                if dfs(vecino):
                    return True
            elif en_recursion[vecino]:
                return True
        
        en_recursion[v] = False
        return False

    for nodo in G.nodes:
        if not visitado[nodo]:
            if dfs(nodo):
                return True
    return False 

```

```python
import networkx as nx
import matplotlib.pyplot as plt

G_dirigido = nx.DiGraph()
G_dirigido.add_edges_from([
    ('A', 'B'),
    ('B', 'C'),
    ('C', 'D'),
    ('D', 'B')  # Esta arista crea un ciclo: B -> C -> D -> B
])

tiene_ciclo = tiene_ciclo_dfs(G_dirigido)
print("¿El grafo tiene ciclo?", tiene_ciclo)

# Graficar el grafo dirigido
plt.figure(figsize=(6, 4))
pos_dir = nx.spring_layout(G_dirigido)
nx.draw(G_dirigido, pos_dir, with_labels=True, node_color='lightblue', edge_color='gray', node_size=400, arrows=True)
plt.title("Grafo dirigido con ciclo")
plt.show()
```

```python
from unittest import TestCase, main

class TestCiclos(TestCase):
    def test_grafo_con_ciclo(self):
        G = nx.DiGraph()
        G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'A')])
        self.assertTrue(tiene_ciclo_dfs(G))

    def test_grafo_sin_ciclo(self):
        G = nx.DiGraph()
        G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D')])
        self.assertFalse(tiene_ciclo_dfs(G))

    def test_grafo_componentes_desconectadas(self):
        G = nx.DiGraph()
        G.add_edges_from([('A', 'B'), ('C', 'D')])  # Sin ciclos
        self.assertFalse(tiene_ciclo_dfs(G))
        G.add_edge('D', 'C')  # Crea un ciclo en la componente C-D
        self.assertTrue(tiene_ciclo_dfs(G))

    def test_grafo_vacio(self):
        G = nx.DiGraph()
        self.assertFalse(tiene_ciclo_dfs(G))

    def test_grafo_un_nodo(self):
        G = nx.DiGraph()
        G.add_node('A')
        self.assertFalse(tiene_ciclo_dfs(G))

    def test_grafo_desconectados_y_ciclo_al_final(self):
        G = nx.DiGraph()
        # Nodos desconectados
        G.add_node('A')
        G.add_node('B')
        # Componente conectada con ciclo
        G.add_edges_from([('X', 'Y'), ('Y', 'Z'), ('Z', 'X')])
        self.assertTrue(tiene_ciclo_dfs(G))

test = main(argv=[""], defaultTest="TestCiclos", verbosity=2, exit=False)
len(test.result.failures) == 0
```
